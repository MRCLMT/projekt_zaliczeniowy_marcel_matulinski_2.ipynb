```python
!pip install spacy -U
!python -m spacy download pl_core_news_sm
import spacy

#zadanie 1
nlp = spacy.load("pl_core_news_sm")
def extraction_persName(tekst):

  persons = []
  doc = nlp(tekst)
  for ent in doc.ents:
    if ent.label_ == "persName":
      a = ent.text
      persons.append(a)
  return persons

df["Persons"] = df["text"].apply(extraction_persName)

df
```

```python
#zadanie 2
def extraction_placeName(tekst):

  places = []
  doc = nlp(tekst)
  for ent in doc.ents:
    if ent.label_ == "placeName":
      b = ent.text
      places.append(b)
  return places

  df["Places"] = df["text"].apply(extraction_placeName)

  df
```

```python
 df["Places"] = df["text"].apply(extraction_placeName)
```

```python
df
```

```python
#zadanie 3
nlp = spacy.load("pl_core_news_sm")
def extraction_orgName(tekst):

  organizations = []
  doc = nlp(tekst)
  for ent in doc.ents:
    if ent.label_ == "orgName":
      c = ent.text
      organizations.append(c)
  return organizations
```

```python
df["Organizations"] = df["text"].apply(extraction_orgName)
```
